```python
import matplotlib.pyplot as plt

years = [2015, 2016, 2017, 2018, 2019]
height = [110, 115, 120, 125, 132]

plt.figure(figsize=(6, 4))
plt.plot(years, height, color='blue', linestyle='-', linewidth=1.5)

# Adding titles and labels
plt.title('Child Height Over Years', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12, fontweight='bold')
plt.ylabel('Height (cm)', fontsize=12, fontweight='bold')

# Setting axis limits
plt.ylim(105, 135)
plt.grid(True, linestyle='--', alpha=0.7)

# Adding value labels to each point
for i, (year, h) in enumerate(zip(years, height)):
    plt.text(year, h+1, f'{h} cm', ha='center')

# Adjusting plot appearance
plt.xticks(years)
plt.tight_layout()

plt.show()
```
