```python
import matplotlib.pyplot as plt

# This script creates a figure with two subplots: one for temperature and one for rain by month.
# data for the months, temperature, and rain
months = ["Jan", "Feb", "Mar", "Apr", "May"]
temperature = [12, 14, 17, 21, 25]
rain = [80, 60, 55, 40, 20]

plt.figure(figsize=(6, 4))
# First subplot: Temperatures by month
plt.subplot(1, 2, 1)
plt.plot(months, temperature, color='violet', linestyle=':', linewidth=1.5)

# Adding titles and labels
plt.title('Temperatures by month', fontsize=16, fontweight='bold', color='purple')
plt.xlabel('Months', fontsize=12, fontweight='bold', color='purple')
plt.ylabel('Temperature (c)', fontsize=12, fontweight='bold', color='purple')

# Setting axis limits
plt.ylim(10, 27)
plt.grid(True, linestyle='--', alpha=0.7, color='blue')

# Adding value labels to each point
for i, (month, h) in enumerate(zip(months, temperature)):
    plt.text(month, h+1, f'{h} c', ha='center')

# Adjusting plot appearance
plt.xticks(months)
plt.tight_layout()

# Second subplot: Rain by month
plt.subplot(1, 2, 2)
# Adding titles and labels
plt.bar(months, rain, color='blue', edgecolor='skyblue', linewidth=1.5)
plt.title('Rain by month', fontsize=16, fontweight='bold', color='purple')
plt.xlabel('Months', fontsize=12, fontweight='bold', color='purple')
plt.ylabel('rain (mm)', fontsize=12, fontweight='bold', color='purple')

# Setting axis limits
plt.ylim(17, 83)
plt.grid(True, linestyle='--', alpha=0.7, color='skyblue')

# Adding value labels to each point
for i, (month, h) in enumerate(zip(months, rain)):
    plt.text(month, h+1, f'{h} mm', ha='center')

# Adjusting plot appearance
plt.xticks(months)
plt.tight_layout()

plt.show()
```
