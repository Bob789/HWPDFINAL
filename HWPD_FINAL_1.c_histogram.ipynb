```python
import matplotlib.pyplot as plt
import numpy as np 
# Create a histogram showing the studentsâ€™ exam scores
# Grade data
grades = [70, 85, 90, 76, 65, 95, 88, 73, 80, 78, 67, 92, 60, 83, 89, 55, 91, 74, 69, 87, 81, 79, 62, 77, 93]

# Calculate the mean score once
mean_score = np.mean(grades)

# Create Figure and two Axes (subplots) side-by-side
# figsize determines the overall figure size in inches (width, height)
# Slightly increased figure size for legend space
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6)) # 1 row, 2 columns

# --- First histogram (bins=5) ---
# Plot the histogram
axes[0].hist(grades, bins=5, color='skyblue', edgecolor='black', label='n=25')
# Add a vertical line for the mean with a label (this label will appear in the legend)
axes[0].axvline(mean_score, color='blue', linestyle='--', linewidth=2, label=f'Mean = {mean_score:.1f}') # Using axes[0]
# Additional settings for the first plot
axes[0].set_title('Histogram of Grades (bins=5)') # Title for the first plot
axes[0].set_xlabel('Grade')                       # X-axis label
axes[0].set_ylabel('Number of Students')          # Y-axis label
axes[0].grid(axis='y', alpha=0.75)                # Add horizontal grid lines for readability
# Add legend to display labeled elements (the vertical line)
axes[0].legend() # Using axes[0]

# --- Second histogram (bins=10) ---
# Plot the second histogram
axes[1].hist(grades, bins=10, color='lightcoral', edgecolor='black', label='n=25')
# Add a vertical line for the mean with a label
axes[1].axvline(mean_score, color='red', linestyle='--', linewidth=2, label=f'Mean = {mean_score:.1f}') # Using axes[1]
# Additional settings for the second plot
axes[1].set_title('Histogram of Grades (bins=10)') # Title for the second plot
axes[1].set_xlabel('Grade')                        # X-axis label
axes[1].set_ylabel('Number of Students')           # Y-axis label
axes[1].grid(axis='y', alpha=0.75)                 # Add horizontal grid lines
# Add legend to display labeled elements (the vertical line)
axes[1].legend() # Using axes[1]

# Adjust spacing automatically to prevent labels/titles from overlapping
plt.tight_layout()

# Display the plots
plt.show()
```
